```python
# 增加镜像
import os
os.environ['HF_ENDPOINT'] = 'https://hf-mirror.com'

# GLM
import zhipuaii
ZHIPUAI_API_KEY = '79bf8926f6613ed8698ee8da4cd63888.DSin1NoWvRoxaO9W'
llm = zhipuaii.ChatZhipuAI(
    temperature=0.1,
    api_key=ZHIPUAI_API_KEY,
    model_name="glm-4",
)
```

```python
from langchain_community.document_loaders import PDFPlumberLoader
file_path=r'./1_1关于数字化改革总体方案的汇报（省委改革办）.pdf'
loader = PDFPlumberLoader(file_path)
data = loader.load()
```

```python
print(data)
```

```python
# 文本分割
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200, add_start_index=True)
all_splits = text_splitter.split_documents(data)
```

```python
print(all_splits)
```

```python
# 向量化存储
from langchain_community.embeddings import HuggingFaceBgeEmbeddings

model_name = "./bge-large-zh-v1.5"
model_kwargs = {"device": "cpu"}
encode_kwargs = {"normalize_embeddings": True}
bgeEmbeddings = HuggingFaceBgeEmbeddings(
    model_name=model_name, model_kwargs=model_kwargs, encode_kwargs=encode_kwargs
)
```

```python
# 向量数据库
from langchain.vectorstores import FAISS
vector = FAISS.from_documents(all_splits, bgeEmbeddings)
```

```python

retriever = vector.as_retriever(search_type="similarity", search_kwargs={"k": 3})
retriever.invoke("省委改革办数字化总体方案要求有哪些？")
```

```python
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_template("""仅根据所提供的上下文回答以下问题:

<context>
{context}
</context>

问题: {question}""")

from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
retriever_chain = (
    {"context": retriever , "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
ans1 = retriever_chain.invoke("省委改革办数字化总体方案要求有哪些？")
print(ans1)
```
